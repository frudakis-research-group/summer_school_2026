# Python Basics

```python
menu_meals = {'pizza': 20.5, 'burger': 18.2, 'salad': 25}
menu_deserts = {'cake': 10.5, 'ice_cream': 20, 'fruit': 5}
menu_drinks = {'soda': 2, 'water': 1.5, 'beer': 4}

print('# Meals')
print(menu_meals)

print('# Deserts')
print(menu_deserts)

print('# Drinks')
print(menu_drinks)

meal = input('Select meal: ')
desert = input('Select desert: ')
drink = input('Select drink: ')

above_25 = input('Is your age above 25 (yes/no)? ')

cost = menu_meals[meal] + menu_deserts[desert] + menu_drinks[drink]

# Check if customer is below 25 and total is above 50
# Apply discount
if above_25 == 'yes' and cost > 50:
    total = cost - (0.2 * cost)
else:
    total = cost

# Print the total amount client must pay
print('Total bill: ', total)
```
